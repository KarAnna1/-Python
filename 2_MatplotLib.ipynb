```python
a = np.arange(12)
A = a.reshape([4, 3])
print(A)
At = A.T
print(At)
B = A.dot(At)
print(B)
B_det = np.linalg.det(B)
print(B_det)
```

Матрица B имеет размер (число строк А)х(число столбцов Аt) = 4х4

Обратная мотрица существует только для тех матриц, для которых определитель не равен нулю. В нашем случае det(B) = 0, то есть для матрицы B не существует обратной

Замечание: последовательность от 0 до 12 содержит 13 значений, такая последовательность не может преобразовываться в матрицу 4х3, поэтому последовательность содержит значения от 0 до 11
